```python
x = np.linspace(0,2*np.pi)
```

```python

plt.plot(x,np.sin(x),'r.',label=r'$\rho+\sin(x)$')
plt.plot(x,np.sin(x),'b-')
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
plt.grid()

plt.legend()
plt.show()
```
