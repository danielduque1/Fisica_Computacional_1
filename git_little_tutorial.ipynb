Git is a widely-used, open-source distributed version control system (DVCS) that tracks changes in files and code during software development. It allows multiple people to collaborate on a project, manage different versions of files, and recover earlier versions at any time. 


Common **Git Commands** and Workflow: A typical Git workflow involves a few core commands:

| Command |	Purpose| 
| :--- | :--- |
|**git init**	| Initializes a new local Git repository in a project folder.| 
|**git clone [url]**	| Creates a local copy of a project that already exists on a remote server.| 
|**git add [file]**	| Stages a change, adding a snapshot of the file to a staging area, ready for the next commit.| 
|**git commit -m "[message]"**	| Saves the staged snapshot permanently to the project history with a descriptive message.| 
|**git status**	| Shows the status of changes as untracked, modified, or staged.| 
|**git pull**	| Fetches the latest changes from the remote repository and merges them into your local branch.| 
|**git push**	| Uploads your local commits to the remote repository (e.g., GitHub, GitLab).| 
|**git branch [name]**	| Creates a new branch for isolated development.| 
|**git merge [branch]**	| Combines changes from a specified branch into the current active branch.| 

For more information, see the full reference guide to Git commands. https://git-scm.com/docs

Offically documentation: https://git-scm.com/

### Example: Start a new repository and publish it to GitHub

First, you will need to create a new repository on GitHub. For more information, see Hello World. Do not initialize the repository with a README, .gitignore or License file. This empty repository will await your code.


```python
# create a new directory, and initialize it with git-specific functions
git init my-repo

# change into the `my-repo` directory
cd my-repo

# create the first file in the project
touch README.md

# git isn't aware of the file, stage it
git add README.md

# take a snapshot of the staging area
git commit -m "add README to initial commit"

# provide the path for the repository you created on github
git remote add origin https://github.com/YOUR-USERNAME/YOUR-REPOSITORY-NAME.git

# push changes to github
git push --set-upstream origin main
```

* Maybe you have to use: git@github.com:YOUR-USERNAME/YOUR-REPOSITORY-NAME.git

* and change: main--> master

### Example: Contribute to an existing repository in Github

note that is a new banch...

```python
# download a repository on GitHub to our machine
# Replace `owner/repo` with the owner and name of the repository to clone
#from http
git clone https://github.com/owner/repo.git
#from SSH: better
git clone git@github.com:owner/repo.git

# change into the `repo` directory
cd repo

# create a new branch to store any new changes
git branch my-branch

# switch to that branch (line of development)
git checkout my-branch

# make changes, for example, edit `file1.md` and `file2.md` using the text editor

# stage the changed files
git add file1.md file2.md

# take a snapshot of the staging area (anything that's been added)
git commit -m "my snapshot"

# push changes to github
git push --set-upstream origin my-branch
```
