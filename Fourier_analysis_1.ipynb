# Fourier analysis 1. A simple behaviour.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's see two examples of discret Fourier analysis.

##Example 1: For the Heaviside function:

$$f(x) = 1,\hspace{1cm} 0 <x < \pi$$

$$f(x) = 2,\hspace{1cm}  \pi <x < 2\pi\,.$$

The Fourier expansion in the basis:  ${\phi_n(x)=\dfrac{1}{\sqrt{2L}}e^{i\,n\pi x/L}}$ en $c\leq x \leq c+2L$, $n=-\infty ... 0 ...+\infty$,
is given by:


\begin{equation}
    f(x)= \dfrac{3}{2} -\dfrac{2}{\pi}\sum_{n=0}^{\infty}\dfrac{\sin((2n+1)x)}{2n+1}
\end{equation}

..see physical mathematica Alonso S. book

```python
#Programing the f(x) function:
def f(n,x):
    sum = 0 #ini the sum in 0
    for i in range(n):
      sum += np.sin((2*i+1.)*x)/(2*i+1.)

    f = (3./2.) -(2./np.pi)*sum
    return f
```

```python
#a = np.arange(0,2*np.pi,0.1)
a = np.linspace(0, 2*np.pi, 100)

plt.hlines(1,np.pi,0)
plt.hlines(2,2*np.pi,np.pi)
plt.vlines(np.pi,1,2,linestyles='dashed')

plt.plot(a, f(1,a), 'r-',label='n=2')

plt.title('Fourier analysis')

plt.xlabel(r'$x$',size=15)
plt.ylabel(r'$f(x)$',size=12)
plt.legend(prop={'size':10},loc='best')

plt.xlim(0,2*np.pi)
plt.show()
```

#Example 2: Funcin valor absoluto:

$$f(x) =  -x,\hspace{1 cm}   -\pi \leq x \leq 0 $$
$$f(x) =  +x,\hspace{1 cm}   0 < x \leq \pi\,. $$

$\textbf{Expandir}$ la funcin $f(x)$ en la base discreta, ortonormal de Fourier: ${\phi_n(x)=\dfrac{1}{\sqrt{2L}}e^{i\,n\pi x/L}}$ en $c\leq x \leq c+2L$, $n=-\infty ... 0 ...+\infty$, del espacio de Hilbert de funciones de cuadrado integrable $\mathscr{L}^2$ con peso $p(x)=1$.

Solution:

$$f(x) = \dfrac{\pi}{2}-\dfrac{2}{\pi}\sum_{k=-\infty, \neq 0}^{+\infty}\dfrac{1}{(2k+1)^2}e^{i(2k+1)x}$$

$$f(x) = \dfrac{\pi}{2}-\dfrac{4}{\pi}\sum_{k=0}^{ \infty} \dfrac{\cos((2k+1)x)}{(2k+1)^2}$$

```python
def xabsolut(n,x):
    ''' Absolut value expansion'''
    sum = np.pi/2

    for i in range(0,n):
      sum += -(4/np.pi)*np.cos((2*i+1)*x)/(2*i+1)**2

    return sum
```

```python
a = np.linspace(-np.pi, np.pi, 50)

plt.figure(figsize=(6,4))
plt.plot(a, np.abs(a),lw=2,label='Numpy')
#plt.plot(a, xabsolut(2,a), lw=1,label='n=2')
plt.plot(a, xabsolut(4,a), lw=1,label='n=4')
#plt.plot(a, xabsolut(6,a), 'r .',label='n=6')
#plt.plot(a, xabsolut(6,a), lw=1, color='red')

plt.title('Fourier analysis')

plt.xlabel(r'$x$',size=15)
plt.ylabel(r'$f(x)$',size=12)
plt.legend(prop={'size':10},loc='best')
plt.grid()

plt.show()
```
